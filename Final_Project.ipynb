## Step-03: Plotting Graphs

### Comparing Pollution & Quality by Country

```python
plt.figure(figsize=(30,15))
sns.barplot(data=df,x='Water Pollution Index',y='Water Quality Score',hue='Country',palette='Set1',saturation=100,edgecolor='black')
plt.title('Compare Pollution & Quality by Country',fontsize=14,fontweight='bold')
plt.tight_layout()
plt.show()
```

### KEY Insights:

#### Inverse Relationship is Clear
- As the Water Pollution Index increases, the Water Quality Score decreases  which is expected and confirms data consistency.
- This trend is visible across all countries.



### Yearly trends of Water Pollution Index	

```python
plt.figure(figsize=(30,15))
sns.lineplot(data=df,x='Year',y='Water Pollution Index',errorbar=None,marker='o',color='green',linestyle='--',markersize=20,)
plt.title('Yearly trends of Water Pollution Index',fontsize=14,fontweight='bold')
plt.tight_layout()
plt.grid()
plt.show()
```

#### Key Insights

1.**High Pollution Early in the Decade (20102012):**

- 2010: ~75.2

- 2011: Increased slightly to ~77.5 (peak)

- 2012: Slight improvement (~73.3), but still high

2.**Improvement Phase (20132014):**

- WPI dropped significantly to ~65.2 (2013) and even further to ~57.5 in 2014

- Suggests successful interventions or temporary improvements (possibly rainfall, regulation, etc.)

3.**Pollution Surge Again (20152016):**

- 2015: Spike back to ~72.6

- 2016: Reaches ~74.8  second-highest point in the timeline

4.**Fluctuating Pattern (20172020):**

- 2017: Sudden drop to ~63.5

- 20182020: Gradual increase again  stabilizes around 7075 by 2020



### Industrial Waste vs. Water Pollution	

```python
plt.figure(figsize=(30,15))
sns.regplot(data=df,x='Industrial Waste (%)',y='Water Pollution Index',marker='o',color='green'
           ,line_kws=dict(color='blue',linestyle='-',linewidth=1),
           scatter_kws=dict(s=200,color='red',alpha=1),)
plt.title('Industrial Waste vs. Water Pollution',fontsize=14,fontweight='bold')
plt.tight_layout()
plt.grid()
plt.show()
```

#### Key Insights

1.**Positive Correlation**

- The regression line slopes upward, indicating a positive relationship between Industrial Waste (%) and Water Pollution Index.
- As industrial waste increases, water pollution also tends to increase.

2.**Regression Line Interpretation**
- The blue regression line gives the trend direction.
- It helps in predicting Water Pollution Index based on given Industrial Waste %.



### Sewage Treatment vs. Pollution

```python
plt.figure(figsize=(30,15))
sns.scatterplot(data=df,x='Sewage Treatment (%)',y='Water Pollution Index',marker='o',s=200,hue='Year')
plt.title('Sewage Treatment vs. Water Pollution',fontsize=14,fontweight='bold')
plt.tight_layout()
plt.grid()
plt.show()
```

#### Key Insights

1.**Inverse Relationship Observed:**

- In general, as Sewage Treatment (%) increases, the Water Pollution Index decreases.

- This suggests that better sewage treatment reduces pollution in water sources.

2.**Yearly Trend Differences:**

- Some years may show better performance even at mid-level treatment percentages.



### Access to Clean Water vs. Population Exposed

```python
plt.figure(figsize=(30,15))
sns.scatterplot(data=df,x='Access to Clean Water (%)',y='Population Exposed',marker='o',s=200,hue='Country',style='Country')
plt.title('Access to Clean Water vs. Population Exposed',fontsize=14,fontweight='bold')
plt.tight_layout()
plt.grid()
plt.show()
```

#### Key Insights

1.**Inverse Relationship Expected:**

- As the Access to Clean Water (%) increases, ideally, the Population Exposed to unsafe water should decrease.


2.**Country-wise Clusters:**


- Countries with higher access and lower exposed population are performing better.

- Countries below the general downward trend may be struggling with equitable water access despite infrastructure.



### Water Diseases Impact

```python
plt.figure(figsize=(15,5))
sns.barplot(data=df,x='Country',y='Mortality Rate from Water Diseases',hue='Country',edgecolor='black',palette='Set3',saturation=2)
plt.title('Water Diseases Impact',fontsize=16,fontweight='bold')
plt.grid(axis='y',linestyle=':')
plt.show()
```

#### Key Insights

- Countries with the highest mortality rates from waterborne diseases indicate poor water quality and weak healthcare systems.

- Countries with low mortality rates likely have better access to clean water, sanitation, and health services.



### Pollution Sources per Country	

```python
sns.countplot(data=df,x='Pollution Source',hue='Country',edgecolor='black',width=0.8)
plt.title('Pollution Source per_Country',fontsize=16)
```

#### Key Insights

- Different countries have different dominant pollution sources, indicating regional variation in industrial activities and waste management practices.

- Some countries contribute to multiple pollution sources, highlighting broader environmental management issues.

- The most frequent pollution sources (bars with highest counts) suggest the primary causes of water pollution across all countriessuch as industrial waste or sewage discharge.

- Countries with higher bar counts across multiple sources may lack strict environmental regulations or enforcement.

#### **Conclusion**
Based on the analysis of the water pollution dataset, it is evident that industrial waste plays a significant role in increasing the water pollution index, with a strong positive correlation observed between industrial discharge and pollution levels. Countries with higher sewage treatment percentages generally show lower pollution levels, highlighting the importance of effective sewage infrastructure. However, inconsistencies in treatment efficiency may still lead to high pollution in some areas. Additionally, access to clean water remains a critical issue, as nations with limited access show higher population exposure to unsafe water sources, directly contributing to increased mortality from waterborne diseases. The data also reveals that pollution sources vary by country, with industrial waste, sewage, and agricultural runoff being common culprits. To combat these challenges, it is essential for governments to enforce stricter industrial waste regulations and invest in sewage treatment systems. Expanding access to clean water, raising public awareness about hygiene and pollution control, and designing country-specific strategies are also crucial. Furthermore, promoting international cooperation for managing transboundary water bodies and adopting data-driven monitoring systems can ensure long-term sustainability and public health protection.
